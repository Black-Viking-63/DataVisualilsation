### Что делает geom_col()? Чем он отличается от geom_bar()?

`geom_col()` представляет значения данных в виде столбцов.<br>
`geom_bar()` делает высоту полосы пропорциональной количеству наблюдений в каждой группе.<br>
то есть `geom_col()` - это альтернативная версия geom_bar, которая выставляет высоту столбцов в зависимости от какой-либо переменной в данных.

```python
#ggplot(data=diamonds) + geom_col(mapping=aes(x="cut", y="clarity"))
```

### Что произойдёт, если

или
ggplot(data=diamonds) +\
geom_bar(mapping=aes(x="cut", fill="color", y="..prop.."))
### запустить без "group=1"? Почему?

`group=1` необходим нам для приведения всех типов к единой группе

```python
ggplot(data=diamonds)+\
geom_bar(mapping=aes(x="cut", y="..prop.."),group=1)
```

```python
ggplot(data=diamonds)+\
geom_bar(mapping=aes(x="cut", y="..prop.."))
```

### Зачем тут coord_fixed()? Что делает geom_abline()?

`coord_fixed()` - задает соотношение сторон для системы координат графика<br>
`geom_abline()` строит диагональную линию с заданным наклоном и сдвигом

```python
ggplot(data=mpg, mapping=aes(x="cty", y="hwy")) + geom_point() + geom_abline() + coord_fixed()
```
