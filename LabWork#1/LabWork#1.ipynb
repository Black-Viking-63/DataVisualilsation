Построение нескольких графиков

```python
import pandas as pd, numpy as np
pd.set_option('display.max_columns', None)
import matplotlib.pyplot as plt, seaborn as sns 
import warnings, sys
if not sys.warnoptions:
    warnings.simplefilter("ignore")
from datetime import datetime 
```

```python
frame = pd.read_csv('merc.csv')
frame.head(10)
```

## размер налога относительно года 

```python
sns.set_style("darkgrid") 
sns.lineplot(data=frame, x="year", y="tax") # 
```

## размер сбора на машину в зависимости от года производства и объема двигателя

```python
pivot = frame.pivot_table(
    index=["year"],
    columns=["tax"],
    values="engineSize",
    aggfunc=np.average)
sns.heatmap(pivot) # 
```

## число машин с популярным типом топлива на вторичном рынке

```python
sns.histplot(frame['fuelType'])
plt.show() #
```

## сводный интерактивный график

```python
import plotly.express as px
px.scatter_matrix(frame[["year", 'price', 'mileage', 'engineSize']])
```
